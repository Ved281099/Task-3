```python
import pywedge as pw
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
```

```python
data=pd.read_csv("SampleSuperstore.csv")
data.head()
```

```python
data.describe()
```

```python
data.info()
```

# Correlation plot

```python
corr=data.corr()
```

```python
sb.heatmap(corr,annot=True);
```

# Sales vs Profit

```python
plt.scatter(data['Sales'],data['Profit'])
plt.xlabel('sales')
plt.ylabel('profit');
```

# Profits per segment

```python
plt.figure(figsize=(7,7))
data["Segment"].value_counts().plot.pie()
plt.show()
```

# Profits per region

```python
plt.figure(figsize=(7,7))
data["Region"].value_counts().plot.pie()
plt.show()
```

# Profit & Sales per state

```python
data.groupby('State')[["Profit","Sales"]].sum().sort_values(by="Sales",ascending=False).plot.bar(color=["red","green"],figsize=(8,5));
```

# Profit per category

```python
data.groupby('Category')[["Profit","Sales"]].sum().sort_values(by="Sales",ascending=False).plot.bar(color=["red","green"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```

# Profit/Loss per sub-category

```python
data.groupby('Sub-Category')[["Profit","Sales"]].sum().sort_values(by="Sales",ascending=False).plot.bar(color=["red","green"],figsize=(8,5))
plt.ylabel("profit/loss and sales")
plt.show()
```

```python
c1=pw.Pywedge_Charts(data,c=None,y='Sales')
```

# Conclusion

We can observe from above scatterplot of profit vs sales that, as number of sales increases profit also increases.So, to make more profit we need to increase our sales.

In correlation plot we can see discount and profit are negatively correlated so, to increase our profit we need to decrease discount and sale & quantity are positively correlated with profit which means we need to increase our sales and quantities of products to obtain more profit.

From above pie charts we observe that profit made from consumer sector is 54.3% and profit made from home office sector is very less that is 17.3%.So, we need to increase our sales in Home office sector to increase our profit in it.Similarly we can see that the profit is higher in East region of country which is around 41.1% and in South & Central region it is 11.2% & 19.8% respectively.So, in South and Central we need to improve our marketing strategies. 

From barplot we can see in states like New York and California we are in surplus but in Colorado state of US we are in deficit. So, we need to work more in Colorado state of US to increase our profit in that state. 

In profit per category bar plot it is observed that profit obtained from office supplies & Technology is more as compared to furniture. So, here we should concentrate on office supplies & technology products only to increase our profit.

Since, from profit/loss barplot we can see that profit obtained by selling copiers,accesories, paper & phones is higher.So, we need to increase quantity of these products and since we are obtaining loss on selling of tables we should decrease quantity of it.
